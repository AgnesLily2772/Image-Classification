```python
from tensorflow.keras.models import load_model
```

```python
import cv2
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
```

```python
loaded_model = load_model("../model/flower_classifier.h5")
```

```python
labels = ['rose', 'sunflower']
```

```python
random_img = cv2.imread("../test/sunflower.jpg")
plt.imshow(random_img)
```

```python
resized_img = tf.image.resize(random_img,(256,256))
```

```python
pred = loaded_model.predict(np.expand_dims(resized_img/255,0))
```

```python
pred[0][0]
```

```python
if(pred[0][0] > 0.5):print("Sunflower")
else:print("Rose")
```
